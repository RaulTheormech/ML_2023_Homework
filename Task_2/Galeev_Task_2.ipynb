# 3. Обучить модели классификации
Для обучения выбраны модели KNN (KNeighborsClassifier)

## 3.0.1 Рассмотрим KNN (KNeighborsClassifier)

```python
# программная реализация алгоритма kNN для классификации
from sklearn.neighbors import KNeighborsClassifier
# программная реализация расчета метрики точности
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
# Создадим объект класса KNN с параметром n_neighbors=3
classifier_kNN = KNeighborsClassifier(n_neighbors=3)

start_time = time.time() # Измерение времени обучения

# Обучение модели
classifier_kNN.fit(X_train, y_train)

end_time = time.time()

training_time = end_time - start_time

# Прогноз
start_time = time.time() # Измерение времени прогноза

y_pred = classifier_kNN.predict(X_test)

end_time = time.time()

predicting_time = end_time - start_time
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print('Точность модели (precision score): ', round(precision*100, 3), '% ', 'recall score: ', round(recall*100, 3), '%', '\nВремя обучения модели: ', round(training_time, 3), ' секунд', 
     '\nВремя прогноза: ', round(predicting_time, 3), ' секунд')
```

## 3.0.2 Рассмотрим наивный Байесовский метод (Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB
```

```python
GaussianNB = GaussianNB()

start_time = time.time()
GaussianNB.fit(X_train, y_train)
end_time = time.time()

training_time = end_time - start_time

start_time = time.time()
y_pred = GaussianNB.predict(X_test)
end_time = time.time()

predicting_time = end_time - start_time
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print('Точность модели (precision score): ', round(precision*100, 3), '% ', 'recall score: ', round(recall*100, 3), '%', '\nВремя обучения модели: ', round(training_time, 3), ' секунд', 
     '\nВремя прогноза: ', round(predicting_time, 3), ' секунд')
```

## 3.0.3 Рассмотрим логистическую регрессию

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Создадим объект класса LogosticRegression
classifier_LR = LogisticRegression(random_state=21, max_iter=1000, multi_class='multinomial')

start_time = time.time()
classifier_LR.fit(X_train, y_train)
end_time = time.time()

training_time = end_time - start_time

start_time = time.time()
y_pred = classifier_LR.predict(X_test)
end_time = time.time()

predicting_time = end_time - start_time
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print('Точность модели (precision score): ', round(precision*100, 3), '% ', 'recall score: ', round(recall*100, 3), '%', '\nВремя обучения модели: ', round(training_time, 3), ' секунд', 
     '\nВремя прогноза: ', round(predicting_time, 3), ' секунд')
```

## 3.0.4 Рассмотрим дерево решений

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier(random_state=21)

start_time = time.time()
DTC.fit(X_train, y_train)
end_time = time.time()

training_time = end_time - start_time

start_time = time.time()
y_pred = DTC.predict(X_test)
end_time = time.time()

predicting_time = end_time - start_time
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print('Точность модели (precision score): ', round(precision*100, 3), '% ', 'recall score: ', round(recall*100, 3), '%', '\nВремя обучения модели: ', round(training_time, 3), ' секунд', 
     '\nВремя прогноза: ', round(predicting_time, 3), ' секунд')
```

## 3.0.5 Рассмотрим алгоритм CatBoostClassifier

```python
from catboost import CatBoostClassifier
```

```python
CB = CatBoostClassifier(verbose=False)

start_time = time.time()
CB.fit(X_train, y_train)
end_time = time.time()

training_time = end_time - start_time

start_time = time.time()
y_pred = CB.predict(X_test)
end_time = time.time()

predicting_time = end_time - start_time
```
